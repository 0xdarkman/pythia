```python
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt
import os

from pythia.core.streams.fpm_time_series import FpmTimeSeries

coins = ["ETH", "LTC", "XRP", "STR", "ETC", "DASH", "XMR", "XEM", "FCT"]
data_dir = R"/home/bernhard/repos/pythia/data"

rates_dir = os.path.join(data_dir, "recordings/poloniex/processed")
data_frames = []
for coin in coins:
    with open(os.path.join(rates_dir, "{}_{}.csv".format("BTC", coin))) as r:
        data_frames.append(pd.read_csv(r))

size = len(data_frames[0])
ax = None
for frame in data_frames:
    frame["index"] = range(1, size + 1)
    ax = frame.plot(kind='line', ax=ax, x='index', y='close')
plt.show()
```

```python
from pythia.core.visualization.assets_view_modeller import AssetViewModeller
import numpy as np

symbol_prices = [f["close"].values for f in data_frames[0:1]]
assets = np.loadtxt(os.path.join(data_dir, R"models/assets-high.csv"), delimiter=',')[:, 0:2]
view = AssetViewModeller(symbol_prices, assets, relative_weights=True)
for prices, weights in view:
    plt.scatter(range(0, size), prices, s=np.clip(weights, 0.01, 1))
    
plt.show()
```
